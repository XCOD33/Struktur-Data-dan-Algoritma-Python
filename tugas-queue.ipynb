```python
from datetime import datetime
import time

class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class Queue:
    def __init__(self, value):
        new_node = Node(value)
        self.first = new_node
        self.last = new_node
        self.length = 1

    def print_queue(self):
        temp = self.first
        while temp is not None:
            print(temp.value)
            temp = temp.next

    def enqueue(self, value):
        new_node = Node(value)
        if self.first is None:
            self.first = new_node
            self.last = new_node
        else:
            self.last.next = new_node
            self.last = new_node
        self.length += 1

    def dequeue(self):
        if self.length == 0:
            return None
        temp = self.first
        if self.length == 1:
            self.first = None
            self.last = None
        else:
            self.first = self.first.next
            temp.next = None
        self.length -= 1
        return temp
    
class This_time:
    def __init__(self):
        self.now = datetime.now()
        self.current_time = self.now.strftime("%H:%M:%S")

current_time = This_time()
email_queue = Queue("rico@ricoaditya.my.id")
arr_email = ["rico@ricoaditya.my.id"]

print(f"Email sukses dikirim ke {arr_email[0]}. waktu : {current_time.current_time}")
print("Alamat email yang baru saja dikirim : ")
email_queue.print_queue()
email_queue.dequeue()

i = email_queue.length
while i <= email_queue.length:
    now = datetime.now()
    current_time = now.strftime("%H:%M:%S")
    
    new_email = input("Masukkan email kamu : ")
    if(new_email == ""):
        break
    email_queue.enqueue(new_email)
    arr_email.append(new_email)
    
    print(f"\nEmail sukses dikirim ke {arr_email[i + 1]}. waktu : {current_time}")
    print("Alamat email yang baru saja dikirim : ")
    email_queue.print_queue()
    email_queue.dequeue()
    i += 1
    email_queue.length += 1
    time.sleep(3)
    if i > email_queue.length:
        break
    
print("\n=== &&& ===")
print("Alamat email yang berhasil dikirimkan : ")
for i in range(len(arr_email)):
    print(f"{i + 1}. {arr_email[i]}")

print("\n=== &&& ===")
print("Alamat email di antrian saat ini : ")
email_queue.print_queue()
```
